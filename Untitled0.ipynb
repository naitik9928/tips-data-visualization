```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head(10)
```

```python
pivot = df.pivot_table(index="day", columns="time",values="tip",aggfunc="mean")
sns.heatmap(pivot)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(x="day",y="tip",data=df,hue="sex",)
sns.set_style("whitegrid")
sns.set_context("talk")

plt.show()
```

```python
g=sns.FacetGrid(df,col="day")
g.map(sns.scatterplot,"total_bill","tip")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in tips dataset
df = sns.load_dataset("tips")

# Basic violin plot
plt.figure(figsize=(8,6))
sns.violinplot(x="day", y="tip", data=df,hue="sex")
plt.title("Violin Plot of Tips by Day")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```
